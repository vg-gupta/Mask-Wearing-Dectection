```python
import cv2
import matplotlib.pyplot as plt

# Path to the detected image (Update this path based on your YOLOv5 output)
img_path = "lab-3\\yolov5\\runs\\detect\\exp2\\0_Concern-In-China-As-Mystery-Virus-Spreads_jpg.rf.5633f5fe7a9b926101b7fc16615dfb6a.jpg"

# Load image using OpenCV
img = cv2.imread(img_path)

# Check if image was loaded properly
if img is None:
    print(f"Error: Image not found at {img_path}")
else:
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert from BGR to RGB
    
    # Plot the image
    plt.figure(figsize=(10, 5))
    plt.imshow(img)
    plt.axis("off")
    plt.title("YOLOv5 Detection Output")
    plt.show()
```
